```python
import requests
import time
import os
from bs4 import BeautifulSoup
```

```python
html_files = [file for file in os.listdir(f"HTMLs_Pitchfork") if file.endswith(".html")]
len(html_files)
```

## Parse each article and copy the required contents to "html_pitchfork_metadata.csv" file.

```python
with open("html_pitchfork_metadata.csv", "w") as fw:
    ####################################################
    # Column names on the first row
    ####################################################
    fw.write("article_title\tarticle_bodytext\tposting_date\tarticle_author\n")

    for file_name in html_files:
        ####################################################
        # Column values starting from the second row
        ####################################################
        with open(f"HTMLs_Pitchfork/{file_name}", "r+b") as fr:
            print(file_name)
            soup = BeautifulSoup(fr.read(), "html.parser")
            
            ######################################################################
            # No title exception handling. Skip the article if title is not found.
            ######################################################################
            if soup.find("h1", {"class": "BaseWrap-sc-TURhJ BaseText-fFzBQt SplitScreenContentHeaderHed-fxVOKs eTiIvU fHnbaL dAXWKO"}) == None:
                continue
            else:
                article_title = soup.find("h1", {"class": "BaseWrap-sc-TURhJ BaseText-fFzBQt SplitScreenContentHeaderHed-fxVOKs eTiIvU fHnbaL dAXWKO"}).text.strip()
            
            ##############################################################################
            # No body text exception handling. Skip the article if body text is not found.
            ##############################################################################
            if soup.find("div", {"class": "body__inner-container"}) == None:
                continue
            else:
                text_list = soup.find_all("div", {"class": "body__inner-container"})
                article_bodytext = ""
                
                # The body text is present in 2 containers. Concatenate the text from each 
                # container in to article_bodytext.
                for t in text_list:
                    article_bodytext = article_bodytext + " " + t.text.strip()
                    
            ####################################################
            # No author exception handling
            ####################################################
            if soup.find("a", {"class": "BaseWrap-sc-TURhJ BaseText-fFzBQt BaseLink-gZQqBA BylineLink-eZnyPI eTiIvU iscnmf iQipSE nZHeQ byline__name-link button"}) == None:
                article_author = ""
            else:
                article_author = soup.find("a", {"class": "BaseWrap-sc-TURhJ BaseText-fFzBQt BaseLink-gZQqBA BylineLink-eZnyPI eTiIvU iscnmf iQipSE nZHeQ byline__name-link button"}).text.strip()
            
            ####################################################
            # No posting date exception handling
            ####################################################
            if soup.find("p", {"class": "BaseWrap-sc-TURhJ BaseText-fFzBQt InfoSliceValue-gSTMso eTiIvU bsGTGn"}) == None:
                posting_date = ""
            else:
                p_list = soup.find_all("p", {"class": "BaseWrap-sc-TURhJ BaseText-fFzBQt InfoSliceValue-gSTMso eTiIvU bsGTGn"})
                
                # There can be up to 3 fields with the same 'p' class and the posting date is 
                # always the last field. So retrieve the text from the last field.
                length = len(p_list)
                
                posting_date = p_list[length - 1].text
            
            ############################################################
            # Remove all possible tabs and replace new lines with space.
            ############################################################
            article_title = article_title.replace("\t", "").replace("\n", " ")
            article_bodytext = article_bodytext.replace("\t", "").replace("\n", " ").replace('.”', '”.').replace(".’", "’.").replace(" .", ". ").replace(".", ". ")
            article_aurthor = article_author.replace("\t", "").replace("\n", " ")
            posting_date = posting_date.replace("\t", "").replace("\n", " ")
                        
            fw.write(f"{article_title}\t{article_bodytext}\t{posting_date}\t{article_author}\n")
```
